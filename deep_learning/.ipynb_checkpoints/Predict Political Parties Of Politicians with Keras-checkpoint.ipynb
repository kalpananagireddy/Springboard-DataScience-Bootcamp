

## Predict political party based on votes

This notebook use a public data set of how US congressmen voted on 17 different issues in the year 1984 and predict their polical party based on their votes alone, using a deep neural network!


```python
import pandas as pd

feature_names = ['party','handicapped-infants', 'water-project-cost-sharing', 
                    'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                    'el-salvador-aid', 'religious-groups-in-schools',
                    'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                    'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                    'education-spending', 'superfund-right-to-sue', 'crime',
                    'duty-free-exports', 'export-administration-act-south-africa']

voting_data = pd.read_csv('house-votes-84.data.txt', na_values=['?'], names=feature_names)
voting_data.head()
```

```python
voting_data.describe()
```

```python
voting_data.dropna(inplace=True)
```

```python
voting_data.describe()
```

```python
voting_data.replace(('y','n'),(1,0),inplace=True)
voting_data.replace(('democrat','republican'), (1,0), inplace=True)
voting_data.head()
```

```python
all_features= voting_data[feature_names].drop('party',axis=1).values
all_class = voting_data['party'].values
print(type(all_class))
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import cross_val_score
```

```python
def create_model():
    
    model= Sequential()
    model.add(Dense(10,activation='relu', input_shape=(16,)))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

estimator =KerasClassifier(build_fn=create_model, epochs=60)
```

```python
cv_scores=cross_val_score(estimator, all_features, all_class, cv=5)
```

```python
cv_scores.mean()
```
